```python
batting['strikeRate'] = batting['strikeRate'].astype('str')
batting['strikeRate'] = batting['strikeRate'].str.replace(',', '')
batting['strikeRate'] = pd.to_numeric(batting['strikeRate'], errors='coerce')
```

```python
plt.figure(figsize=(10,10))
sns.barplot(x='current_innings',y='strikeRate',data=batting)
plt.title("barplot for current_innings vs strikeRate")
```

### Observation : We can see her newzealand strikeRate is good in world cup & Australia is in second as strikeRate
